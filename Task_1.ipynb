```python
#Importing the necessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Importing the dataset
dataset = pd.read_csv("https://raw.githubusercontent.com/AdiPersonalWorks/Random/master/student_scores%20-%20student_scores.csv")
x = dataset.iloc[:, :-1]
y = dataset.iloc[:, -1]
x
```

```python
#plotting the distribution of scores
dataset.plot(x = 'Hours', y = 'Scores', style = '.')
plt.title('Hours vs Scores')
plt.xlabel('Hours of studying')
plt.ylabel('Scores')
plt.show()
```

```python
#Splitting the data into training and test set with 80% ratio
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 0)
```

```python
#Now, training the Simple Linear Regression model on the Training Set
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x_train, y_train)
```

```python
#Visualising the training set results
plt.scatter(x_train, y_train, color = 'blue')
plt.plot(x_train, model.predict(x_train), color = 'green')
plt.title('Hours vs Scores(Training Set')
plt.xlabel("Hours of studying")
plt.ylabel('Scores')
plt.show()
```

```python
#Printing the predicted values
y_pred = model.predict(x_test)
y_pred
```

```python
#Finally, we predict the score on basis of hours of studying
hours = 9.5
print("Studying hours is", hours)
print("The predicted score is: ", model.predict([[hours]]))
```
